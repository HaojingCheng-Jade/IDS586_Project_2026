# Clean Minimally


```python
ff = pd.read_csv("../data/Raw/F-F_Research_Data_5_Factors_2x3.csv", skiprows=3)

# rename date column
ff = ff.rename(columns={"Unnamed: 0": "DateKey"})

# keep only numeric rows (monthly data)
ff = ff[pd.to_numeric(ff["DateKey"], errors="coerce").notna()]

# convert to int
ff["DateKey"] = ff["DateKey"].astype(int)

# keep only YYYYMM format (6 digits)
ff = ff[ff["DateKey"] > 100000]

print(ff.tail())
print(ff.shape)
```

```python
# duplicates
print("Duplicates:", df.duplicated(["PERMNO", "MthCalDt"]).sum())

# missing
print(df[["MthRet", "MthPrc", "MthCap"]].isnull().sum())

# basic sanity
print(df["MthRet"].describe())
```

## Drop missing and duplicates

```python
# sort first
df = df.sort_values(["PERMNO", "MthCalDt"])

# drop exact duplicates (keep first)
df = df.drop_duplicates(subset=["PERMNO", "MthCalDt"], keep="first")

print("Duplicates after:", df.duplicated(["PERMNO", "MthCalDt"]).sum())
print("New shape:", df.shape)
```

```python
df = df.dropna(subset=["MthRet", "MthPrc", "MthCap"])

print("Shape after NA drop:", df.shape)
```

## Drop Noise (penny stocks and low value stocks)

```python
df["MthPrc_Abs"] = df["MthPrc"].abs()
df = df[df["MthPrc_Abs"] >= 5]

print(df.shape)
```

## Create DateKey

```python
df["MthCalDt"] = pd.to_datetime(df["MthCalDt"])
df["DateKey"] = df["MthCalDt"].dt.year * 100 + df["MthCalDt"].dt.month

print(df[["MthCalDt", "DateKey"]].head())
```

## Validate FF

```python
# duplicates
print("Duplicates:", ff.duplicated(["DateKey"]).sum())

# missing
print(ff.isnull().sum())

# basic stats
print(ff.describe())
```

```python
cols = ["Mkt-RF", "SMB", "HML", "RMW", "CMA", "RF"]

ff[cols] = ff[cols].apply(pd.to_numeric, errors="coerce")
ff[cols] = ff[cols] / 100

print(ff.dtypes)
```

```python
df_merged = df.merge(ff, on="DateKey", how="inner")

print("Shape:", df_merged.shape)
```

```python
print(df_merged[["DateKey", "MthRet", "Mkt-RF", "SMB"]].head())

print("Unique stocks per month (sample):")
print(df_merged.groupby("DateKey")["PERMNO"].nunique().head())
```

## Target Feature Construction



```python
df_merged["ExcessRet"] = df_merged["MthRet"] - df_merged["RF"]

df_merged[["MthRet", "RF", "ExcessRet"]].head()
```

```python
# Winsorize Excess Returns (1% / 99%)
lower = df_merged["ExcessRet"].quantile(0.01)
upper = df_merged["ExcessRet"].quantile(0.99)

df_merged["ExcessRet"] = df_merged["ExcessRet"].clip(lower, upper)
```

```python
df_merged = df_merged.sort_values(["PERMNO", "MthCalDt"])

df_merged["y_next"] = df_merged.groupby("PERMNO")["ExcessRet"].shift(-1)

df_merged[["PERMNO", "MthCalDt", "ExcessRet", "y_next"]].head(10)
```

```python
# Remove last observation

df_merged = df_merged.dropna(subset=["y_next"])

print(df_merged.shape)
```

```python
df_merged["ret_lag1"] = df_merged.groupby("PERMNO")["ExcessRet"].shift(1)
df_merged[["PERMNO", "MthCalDt", "ExcessRet", "ret_lag1"]].head(10)
```

```python
# drop NaN created by lag

df_merged = df_merged.dropna(subset=["ret_lag1"])
print(df_merged.shape)
```

```python
df_merged["ret_lag2"] = df_merged.groupby("PERMNO")["ExcessRet"].shift(2)
df_merged["ret_lag3"] = df_merged.groupby("PERMNO")["ExcessRet"].shift(3)

df_merged = df_merged.dropna(subset=["ret_lag2", "ret_lag3"])

print(df_merged.shape)
```

```python
df_merged["history_len"] = df_merged.groupby("PERMNO").cumcount() + 1

print(df_merged[["PERMNO", "MthCalDt", "history_len"]].head(10))
```

```python
# filter out those with less history

df_merged = df_merged[df_merged["history_len"] >= 36]

print(df_merged.shape)
```

```python
# add size feature
import numpy as np

df_merged["log_size"] = np.log(df_merged["MthCap"])
```
